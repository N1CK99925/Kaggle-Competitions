```python
train = train.drop(columns=['Name','Ticket'])
test = test.drop(columns=['Name','Ticket'])
```

```python
#Feaatures

x_train = train.drop('Survived',axis=1)
y_train = train['Survived']
```

```python
x_test = test[x_train.columns]
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np

best_oob_score = 0
best_param = {}

n_estimators_list = [50,100,200]
max_depth_list = [3,5,10,None]
max_features_list = ['sqrt','log2']
min_leaf_sample_list = [1,2,5,8]
min_sample_split_list = [5,8,10]
```

```python
print("Parameter tuning with OOB : ")
for n_est in n_estimators_list:
    for depth in max_depth_list:
        for feat in max_features_list:
            for leaf in min_leaf_sample_list:
                for split in min_sample_split_list:
                    model = RandomForestClassifier(
                        n_estimators=n_est,
                        max_depth=depth,
                        max_features=feat,
                        min_samples_leaf=leaf,
                        min_samples_split=split,
                        oob_score=True,
                        random_state=1,
                        n_jobs=-1,
                        # warm_start=True # This is nice feature cuz everytime u fit it does make forest again so slow
                    )
                    model.fit(x_train,y_train)
                    #oob score
                    current_oob_score = model.oob_score_
                    print(f"n_est: {n_est:3d} | depth: {str(depth):4} | min_leaf: {leaf:3d} | min_split: {split:3d} | max_feat: {str(feat):5} | OOB: {current_oob_score:.4f}")
                    if current_oob_score > best_oob_score:
                        best_oob_score = current_oob_score
                        best_params = {'n_estimators': n_est, 'max_depth': depth, 'max_features': feat, 'min_samples_leaf': leaf,
                        'min_samples_split':split}
                        best_model = model  # saves thghe best model


print("\n" + "="*50)
print(f"Best OOB Score: {best_oob_score:.4f}")
print(f"Best Parameters: {best_params}")

```

```python
predictions = best_model.predict(x_test)

output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
